```python
for col in ['Teacher_Quality', 'Parental_Education_Level', 'Distance_from_Home']:
    students_df[col].fillna(students_df[col].mode()[0], inplace=True)

students_df.isnull().sum()
```

```python
cat_cols = students_df.select_dtypes(include='object').columns

for col in cat_cols:
    print(students_df[col].value_counts())
    print("\n")
```

```python
ordinal_cols = [
    'Parental_Involvement',
    'Access_to_Resources',
    'Motivation_Level',
    'Family_Income',
    'Teacher_Quality',
    'Parental_Education_Level',
    'Distance_from_Home'
]

nominal_cols = [
    'Extracurricular_Activities',
    'Internet_Access',
    'School_Type',
    'Peer_Influence',
    'Learning_Disabilities',
    'Gender'
]
```

```python
for col in ordinal_cols:
    print(col, students_df[col].unique())
```

```python
mapping = {
    'Low': 1,
    'Medium': 2,
    'High': 3
}

for col in [
    'Parental_Involvement',
    'Access_to_Resources',
    'Motivation_Level',
    'Family_Income',
    'Teacher_Quality'
]:
    students_df[col] = students_df[col].map(mapping)
```

```python
students_df['Distance_from_Home'] = students_df['Distance_from_Home'].map({
    'Near': 1,
    'Moderate': 2,
    'Far': 3
})
```

```python
students_df['Parental_Education_Level'] = students_df['Parental_Education_Level'].map({
    'High School': 1,
    'College': 2,
    'Postgraduate': 3
})
```

```python
for col in nominal_cols:
    print(col, students_df[col].unique())
```

```python
binary_map = {
    'Yes': 1,
    'No': 0
}

students_df['Internet_Access'] = students_df['Internet_Access'].map(binary_map)
students_df['Extracurricular_Activities'] = students_df['Extracurricular_Activities'].map(binary_map)
students_df['Learning_Disabilities'] = students_df['Learning_Disabilities'].map(binary_map)
```

```python
students_df['Gender'] = students_df['Gender'].map({"Male" : 1, "Female" : 0})
```

```python
students_df['School_Type'] = students_df['School_Type'].map({
    'Public': 0,
    'Private': 1
})
```

```python
students_df['Peer_Influence'] = students_df['Peer_Influence'].map({
    'Negative': 0,
    'Neutral': 1,
    'Positive': 2
})
```

```python
students_df.info()
```

```python
students_df['Parental_Education_Level'] = students_df['Parental_Education_Level'].astype(int)
students_df['Distance_from_Home'] = students_df['Distance_from_Home'].astype(int)
```

```python
students_df.rename(columns={
    'Parental_Involvement': 'parental_involvement_id',
    'Access_to_Resources': 'access_resource_id',
    'Extracurricular_Activities': 'extracurricular_activity_id',
    'Motivation_Level': 'motivation_id',
    'Internet_Access': 'internet_access_id',
    'Family_Income': 'family_income_id',
    'Teacher_Quality': 'teacher_quality_id',
    'School_Type': 'school_type_id',
    'Peer_Influence': 'peer_influence_id',
    'Learning_Disabilities': 'learning_disability_id',
    'Parental_Education_Level': 'parental_education_id',
    'Distance_from_Home': 'distance_from_home_id',
    'Gender': 'gender_id'
}, inplace=True)
```
